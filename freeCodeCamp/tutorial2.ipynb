1. Import the required class from PySpark to create a Spark session.

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
```

2. Write the statement to create a SparkSession with the application name "Dataframe".

```python
spark = SparkSession.builder.appName("Dataframe").getOrCreate()
```

3. Display the SparkSession object in the Jupyter Notebook.

```python
spark
```

4. Write the command to read a CSV file into a PySpark DataFrame using Spark.

```python
df = spark.read.csv("test1.csv",header=True,inferSchema=True)
```

5. How do you display the contents of the PySpark DataFrame?

```python
df.show()
```

6. Write the statement used to select specific columns (Name and Experience) from the DataFrame and display them.

```python
df.select("Name","Experience").show()
```

7. Write the syntax used to access a single column (Name) from the PySpark DataFrame.

```python
df.select("Name").show()
```

8. How do you check the data types of all columns in a PySpark DataFrame?

```python
df.dtypes
```

9. Write the command used to generate descriptive statistics of the DataFrame, similar to Pandas describe().

```python
df.describe().show()
```

10. How do you add a new column to the existing PySpark DataFrame?

```python
df_new = df.withColumn("Experience + 2",df["Experience"]+2)
df_new.show()
```

11. Write the statement used to drop a column from the PySpark DataFrame.

```python
df_new_2 = df.drop("Experience + 2")
df_new_2.show()
```

12. How can you rename an existing column in a PySpark DataFrame?

```python
df_renamed = df_new_2.withColumnRenamed("Name","New Name")
```

13. Write the command used to display the updated DataFrame after column operations.

```python
df_renamed.show()
```
