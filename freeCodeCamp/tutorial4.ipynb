**1. Import the required class from PySpark to create a SparkSession.**

```python
from pyspark.sql import SparkSession
```

**2. Write the statement to create a SparkSession with a suitable application name.**

```python
spark = SparkSession.builder.appName("Tutorial4").getOrCreate()
```

**3. Display the SparkSession object in the Jupyter Notebook.**

```python
spark
```

**4. Write the command used to read a CSV file into a PySpark DataFrame with the header option enabled.**

```python
df = spark.read.csv("test1.csv", header=True, inferSchema=True)
```

**5. Display the contents of the PySpark DataFrame after loading the dataset.**

```python
df.show()
```

**6. Write the statement used to filter rows where a column value is greater than a specific number.**

```python
df.filter(df["age"] > 25).show()
```

**7. How do you filter rows where a column value is less than or equal to a given value?**

```python
df.filter(df["Salary"] <= 20000).show()
```

**8. Write the PySpark statement to filter rows based on equality condition.**

```python
df.filter(df["Salary"] == 20000).show()
```

**9. How do you apply multiple filter conditions using logical AND (&) in PySpark?**

```python
df.filter((df["Salary"] <= 20000) & (df["age"] > 25)).show()
```

**10. Write the statement used to apply multiple filter conditions using logical OR (|).**

```python
df.filter((df["Salary"] > 20000) | (df["age"] > 25)).show()
```

**11. How can you filter rows using a NOT condition in PySpark?**

```python
df.filter(~(df["Salary"] == 20000)).show()
```

**12. Write the command used to filter rows based on a string column value.**

```python
df.filter(df["Name"].contains("S")).show()
```

**13. How do you select specific columns after applying a filter condition?**

```python
df_1 = df.filter(df["Name"].contains("S")).select("Name", "age")
df_1.show()
```
