**1. Import the required class from PySpark to create a SparkSession.**

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
```

**2. Write the statement to create a SparkSession with an appropriate application name.**

```python
spark = SparkSession.builder.appName("Tutorial5").getOrCreate()
```

**3. Display the SparkSession object in the Jupyter Notebook.**

```python
spark
```

**4. Write the command used to read a CSV file into a PySpark DataFrame with the header option enabled.**

```python
df = spark.read.csv("test3.csv", header=True, inferSchema=True)
```

**5. Display the contents of the DataFrame after loading the dataset.**

```python
df.show()
```

**6. Write the statement used to group the DataFrame by a single column.**

```python
df_grouped = df.groupBy("Departments")
df_grouped
```

**7. How do you calculate the sum of a numeric column using groupBy()?**

```python
df_grouped.sum("Salary").show()
```

**8. Write the command used to calculate the mean (average) of a column for each group.**

```python
df_grouped.mean("Salary").show()
```

**9. How do you count the number of rows for each group using PySpark?**

```python
df_grouped.count().show()
```

**10. Write the statement used to calculate the maximum value of a column after grouping.**

```python
df_grouped.max("Salary").show()
```

**11. Write the command used to calculate the minimum value of a column after grouping.**

```python
df_grouped.min("Salary").show()
```

**12. How do you apply multiple aggregate functions on a grouped DataFrame?**

```python
df_grouped_2 = df_grouped.agg(sum("Salary"), mean("Salary"), count("Salary"), max("Salary"), min("Salary"))
df_grouped_2.show()
```

**13. Write the statement used to rename aggregate output columns after performing groupBy operations.**

```python
df_grouped_3 = df_grouped_2.withColumnRenamed("sum(Salary)", "Total Salary")
df_grouped_3 = df_grouped_3.withColumnRenamed("avg(Salary)", "Average Salary")
df_grouped_3 = df_grouped_3.withColumnRenamed("count(Salary)", "Number of Employees")
df_grouped_3 = df_grouped_3.withColumnRenamed("max(Salary)", "Maximum Salary")
df_grouped_3 = df_grouped_3.withColumnRenamed("min(Salary)", "Minimum Salary")
```

**14. How do you display the aggregated DataFrame output?**

```python
df_grouped_3.show()
```
